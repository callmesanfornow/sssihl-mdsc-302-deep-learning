```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plotting Activation Functions

We have plotted, in this notebook:

1. `sigmoid(x)` : **$Ïƒ(x)$**

2.  `tanh(x)` : **$\tanh(x)$**

3. `identity(x)` : **$f(x) = x$**

```python
def sigmoid(x, scale=1):
    return 1 / (1 + np.exp(-scale * x))

def tanh(x, scale=1):
    return scale * np.tanh(x)

def identity(x):
    return x
```

```python
# Generating data for plotting
x = np.linspace(-5, 5, 1000)
y_sigmoid = sigmoid(x, scale=5)
y_tanh = tanh(x, scale=2)
y_identity = identity(x)
```

```python
plt.plot(x, y_sigmoid, label='Sigmoid')
plt.plot(x, y_tanh, label='Tanh')
plt.plot(x, y_identity, label='Identity')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Activation Functions')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
